```python
import os
import glob
import unicodedata

## NFC 방식 - 한글을 유니코드로 표현할 때 글자 단위로 저장

for filename in glob.glob('*.csv'):
  nfc_filename = unicodedata.normalize('NFC', filename)
  os.rename(filename, nfc_filename)
```

```python
##  read_csv 함수 이용해서 데이터 프레임 만들기, 인코딩 형식 정해주

df = pd.read_csv("/content/남산도서관 장서 대출목록 (2021년 04월).csv", encoding = 'euc-kr')
```

```python
## low_momeory 옵션추가 , CSV 파일은 텍스트 파일이라 각 열이 숫자형인지, 문자열인지에 대한 정보가 없음 Pandas는 데이터를 읽을 때 각 열의 데이터 타입을 스스로 추측해야함.
## low_memory=False:
## 파일 전체를 한 번에 메모리에 올린 후 타입을 추측 메모리는 많이 사용하지만, 전체 데이터를 보고 타입을 결정하므로 훨씬 정확하고 경고가 발생하지 않는다는 특징

df = pd.read_csv("/content/남산도서관 장서 대출목록 (2021년 04월).csv", encoding = 'euc-kr', low_memory=False)
```

```python
## 데이터프레임이 정상적으로 생성되었는지 확인
## pandas head()를 사용해서 상위 n개확인 () 가 공란일 경우 기본 5개 조회가능

df.head()
```

```python
## 인코딩 방식, 컬럼 별로 dtype 지정해서 불러오기

df = pd.read_csv('남산도서관 장서 대출목록 (2021년 04월).csv', encoding='euc-kr',
                 dtype={'ISBN': str, '세트 ISBN': str, '주제분류번호': str})
df.head()
```

```python
## 완성된 데이터 프레임을 csv형식으로 저장
df.to_csv('ns_202104.csv')
```

```python
## 인덱스 3번쨰 까지반 반복해서읽어오

with open('ns_202104.csv') as f:
    for i in range(3):
        print(f.readline(), end='')
```
