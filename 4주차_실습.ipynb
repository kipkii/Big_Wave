###XML 문자열을 파이썬 객체로 변환하기: fromstring( ) 함수


```python
x_str = """
<book>
  <name>혼자 공부하는 데이터 분석 </name>
  <author>박해선</author>
  <year>2022</year>
</book>
"""
```

```python
import xml.etree.ElementTree as et
book = et.fromstring(x_str)
```

```python
print(book)
print(type(book))
print(book.tag)
```

###자식 엘리먼트 확인하기: findtext( ) 메서드


```python
book_childs = list(book)
print(book_childs)
```

```python
name, author, year = book_childs
print(name.text)
print(author.text)
print(year.text)
```

```python
name = book.findtext('name')
author = book.findtext('author')
year = book.findtext('year')

print(name)
print(author)
print(year)
```

```python
x2_str = """
<books>

  <book>
    <name>혼자 공부하는 데이터 분석 </name>
    <author>박해선</author>
    <year>2022</year>
  </book>

  <book>
    <name>혼자 공부하는 머신러닝+딥러닝 </name>
    <author>박해선</author>
    <year>2020</year>
  </book>

</books>
"""
```

```python
## fromstring() 이용해 부모 엘리먼트 확인하기
import xml.etree.ElementTree as e

books = et.fromstring(x2_str)
print(books.tag)
```

```python
## 여러 개의 자식 엘리먼트 확인하기: findall( ) 메서드와 for 문
## 동일 이름의 여러자식 노드 찾을떄는 findall() 메서드,for문 같이 사용

for book in books.findall('book'):
  name = book.findtext('name')
  author = book.findtext('author')
  year = book.findtext('year')

  print(name)
  print(author)
  print(year)
  print()
```
